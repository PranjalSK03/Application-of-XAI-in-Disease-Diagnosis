```python
import os

MODE = 0o666

# replace it with the path where your "dataset" folder is :-
PARENT_DIR = "D:/TU STUDIES related/SEMESTER 7/Project_Material"

def create_folder(DIR_PATH, folder_name, MODE):
    
    ''' create a new folder whose path - DIR_PATH/folder_name '''
    
    if not os.path.exists(DIR_PATH + "/" + folder_name):
        path = os.path.join(DIR_PATH, folder_name) 
        os.mkdir(path, MODE)
        print(DIR_PATH + "/" + folder_name , "created successfully")
        
home_dir = "breast_dataset"
create_folder(PARENT_DIR, home_dir, MODE)

DATASET_DIR = PARENT_DIR + "/" + home_dir
sub_dir = ["benign", "malignant", "normal"]
sub_sub_dir = ["image", "mask"]

for dir_name in sub_dir:
    create_folder(DATASET_DIR, dir_name, MODE)
    SUB_DIR = DATASET_DIR + "/" + dir_name
    for each_dir_name in sub_sub_dir:
       create_folder(SUB_DIR, each_dir_name, MODE)
    
```

```python
import cv2

def image_saver(PATH, class_name, sub_sub_dir, image):
    
    ''' function reads the image from the path and save the images to the respective folder'''

    path = PATH + "/" + image
    img = cv2.imread(path, cv2.IMREAD_UNCHANGED)
    DIR = PARENT_DIR + "/" + home_dir + "/" + class_name + "/" + sub_sub_dir
    if(not os.getcwd() == DIR ):
        os.chdir(DIR)
    if sub_sub_dir == 'mask':
        image_name = image.replace('_mask', "")
        cv2.imwrite(image_name, img) 
    else:
        cv2.imwrite(image, img)
    print(image , "created")
    print("\n")
    

DATA_PATH = PARENT_DIR + "/dataset/breast ultrasound/Dataset_BUSI_with_GT"

classes = os.listdir(DATA_PATH)

for each_class in classes:
    CLASS_PATH = DATA_PATH + "/" + each_class
    class_images = os.listdir(CLASS_PATH)
    print(class_images)
    for each_image in class_images:
        if "mask" not in each_image:
            image_saver(CLASS_PATH, each_class, "image", each_image)
        else:
            image_saver(CLASS_PATH, each_class, "mask", each_image)
```

### MERGING SIMILAR IMAGES

```python

def merge_masks(image, image_list, PATH):
    
    main_mask_path = PATH + "/" + image
    main_img = cv2.imread(main_mask_path, cv2.IMREAD_UNCHANGED)
    
    for each_mask in image_list:
        if not image == image_list: 
            mask_path  = PATH + "/" + each_mask
            img = cv2.imread(mask_path, cv2.IMREAD_UNCHANGED)
            
            #incase if dimension different report it
            if not main_img.shape == img.shape:
                print(each_mask + " & " + image +" - The mask images must have the same dimensions.")
                print("main image shape - ", main_img.shape)
                print("other image shape - ", img.shape)
                return
            
            main_img = cv2.bitwise_or(main_img, img)
            
            if os.path.exists(mask_path):
                os.remove(mask_path)
                
    #if the main_image exists then delete it       
    if os.path.exists(main_mask_path):
        os.remove(main_mask_path)  
    
    #saving the new image
    if(not os.getcwd() == PATH ):
        os.chdir(PATH)
    cv2.imwrite(image, main_img)
            
    

for each_folder in sub_dir:
    MASK_PATH = DATA_PATH + "/" + each_folder + "/" + sub_sub_dir[1]
    mask_images = os.listdir(MASK_PATH)
    for each_img in mask_images:
        image_name = each_img.split('.')
        #finding all the images with the same name as main image
        result_strings = [s for s in mask_images if s.startswith(image_name[0]) and s.endswith(".png")]
        if len(result_strings) > 1:
            print(result_strings)
            merge_masks(each_img, result_strings, MASK_PATH)
    

```
