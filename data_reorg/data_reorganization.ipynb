```python
import os

MODE = 0o666

# replace it with the path where your "dataset" folder is :-
PARENT_DIR = "D:/TU STUDIES related/SEMESTER 7/Project_Material"


def create_folder(DIR_PATH, folder_name, MODE):
    
    ''' create a new folder whose path - DIR_PATH/folder_name '''
    
    if not os.path.exists(DIR_PATH + "/" + folder_name):
        path = os.path.join(DIR_PATH, folder_name) 
        os.mkdir(path, MODE)
        print(PARENT_DIR + "/" + folder_name , "created successfully")

#creating home directory
home_dir = "heart_combined_dataset"
create_folder(PARENT_DIR, home_dir, MODE)

#getting inside the home_directory
DATASET_DIR = PARENT_DIR + "/" + home_dir
sub_dir = ["train", "test"]
#creating sub directories train and test
for dir_name in sub_dir:
    create_folder(DATASET_DIR, dir_name, MODE)

#the train sub directory
TRAIN_DIR = DATASET_DIR + "/" + sub_dir[0]
sub_sub_dir = ["image", "mask"]
#creating the folder for the image and mask images
for dir_name in sub_sub_dir:
    create_folder(TRAIN_DIR, dir_name, MODE)
```

Above code block helps to create the required folder structure as required.

```python
import cv2
from matplotlib import pyplot as plt

#The following code is to combine all the image data in different folders and bring it to a single folder

def image_saver(PATH, image_list, sub_sub_dir, ctr):
    
    ''' function reads the image from the various path and save the images to the respective folder'''
    
    for each in image_list:
        path = PATH + "/" + each
        img = cv2.imread(path, cv2.IMREAD_UNCHANGED)
        DIR = PARENT_DIR + "/" + home_dir + "/" + sub_dir[0] + "/" + sub_sub_dir
        if(not os.getcwd() == DIR ):
            os.chdir(DIR)
        cv2.imwrite("" + str(ctr) + ".png", img) 
        print(sub_sub_dir, ctr, "created")
        print("\n")
        ctr += 1
        
    return ctr


TRAIN_DATA_PATH = PARENT_DIR + "/dataset/CTheartdiseasekaggle/data/train"

train_folder_list = os.listdir(TRAIN_DATA_PATH)

image_ctr = 1
mask_ctr = 1

for each_folder in train_folder_list:
    
    IMAGE_PATH = TRAIN_DATA_PATH + "/" + each_folder + "/image"
    MASK_PATH = TRAIN_DATA_PATH + "/" + each_folder + "/mask"
    print(IMAGE_PATH, MASK_PATH)
    images = os.listdir(IMAGE_PATH)
    image_ctr = image_saver(IMAGE_PATH, images, sub_sub_dir[0], image_ctr)
    masks = os.listdir(MASK_PATH)
    mask_ctr = image_saver(MASK_PATH, masks, sub_sub_dir[1], mask_ctr)
```

```python
# the following code helps to convert all the .dcm images to .png images and combine all different folders to a single one
import pydicom as pyd
import numpy as np

#directory to finaly save the png images
DIR = PARENT_DIR + "/" + home_dir + "/" + sub_dir[1]

TEST_DATA_PATH = PARENT_DIR + "/dataset/CTheartdiseasekaggle/data/test"

test_folder_list = os.listdir(TEST_DATA_PATH)

ctr = 1

for each_folder in test_folder_list:
    NEW_PATH = TEST_DATA_PATH + "/" + each_folder
    folder_list = os.listdir(NEW_PATH)
    for folder in folder_list:
        FINAL_PATH = NEW_PATH + "/" + folder
        image_list = os.listdir(FINAL_PATH)
        for each_image in image_list:
            img = pyd.dcmread(FINAL_PATH + "/" + each_image)
            image_pixels = img.pixel_array.astype(float)
            rescaled_image = (np.maximum(image_pixels,0)/image_pixels.max())*255
            final_image = np.uint8(rescaled_image)
            if(not os.getcwd() == DIR ):
                os.chdir(DIR)
            cv2.imwrite(""+ str(ctr) + ".png", final_image) 
            ctr += 1 
            
```
